```python
# Import Modules

import random
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define Variables

homescore    = {}
awayscore    = {}
homescoresum = 0
awayscoresum = 0
homeplayersdata = [[i] * 10 for i in range(1, 10)]
awayplayersdata = [[i] * 10 for i in range(1, 10)]
homeplayerdata = {}
awayplayerdata = {}
scoretmp     = 0
score        = 0
outcount     = 0
inning       = 1
inningflg    = 'top'
home_no      = 0
away_no      = 0
resultlist   = ['HR', '3B', '2B', '1B', 'FB', 'DB', 'OUT']
base1        = False
base2        = False
base3        = False
```

```python
# def Functions

def result_status(result, base1, base2, base3, scoretmp):
    if result == 'HR':
        if base1 == False and base2 == False and base3 == False: #0塁
            scoretmp = scoretmp + 1
            base1 = False
            base2 = False
            base3 = False
                    
        elif base1 == True  and base2 == False and base3 == False: #1塁
            scoretmp = scoretmp + 2
            base1 = False
            base2 = False
            base3 = False
                    
        elif base1 == False and base2 == True  and base3 == False: #2塁
            scoretmp = scoretmp + 2
            base1 = False
            base2 = False
            base3 = False
                    
        elif base1 == False and base2 == False and base3 == True:  #3塁
            scoretmp = scoretmp + 2
            base1 = False
            base2 = False
            base3 = False
                    
        elif base1 == True  and base2 == True  and base3 == False: #1-2塁
            scoretmp = scoretmp + 3
            base1 = False
            base2 = False
            base3 = False
                    
        elif base1 == True  and base2 == False and base3 == True:  #1-3塁
            scoretmp = scoretmp + 3
            base1 = False
            base2 = False
            base3 = False
                    
        elif base1 == False and base2 == True  and base3 == True:  #2-3塁
            scoretmp = scoretmp + 3
            base1 = False
            base2 = False
            base3 = False
                    
        elif base1 == True  and base2 == True  and base3 == True:  #1-2-3塁
            scoretmp = scoretmp + 4
            base1 = False
            base2 = False
            base3 = False
                    
    elif result == '3B':
        if base1 == False and base2 == False and base3 == False: #0塁
            scoretmp = scoretmp + 0
            base1 = False
            base2 = False
            base3 = True
                    
        elif base1 == True  and base2 == False and base3 == False: #1塁
            scoretmp = scoretmp + 1
            base1 = False
            base2 = False
            base3 = True
                    
        elif base1 == False and base2 == True  and base3 == False: #2塁
            scoretmp = scoretmp + 1
            base1 = False
            base2 = False
            base3 = True
                    
        elif base1 == False and base2 == False and base3 == True:  #3塁
            scoretmp = scoretmp + 1
            base1 = False
            base2 = False
            base3 = True
                    
        elif base1 == True  and base2 == True  and base3 == False: #1-2塁
            scoretmp = scoretmp + 2
            base1 = False
            base2 = False
            base3 = True
                    
        elif base1 == True  and base2 == False and base3 == True:  #1-3塁
            scoretmp = scoretmp + 2
            base1 = False
            base2 = False
            base3 = True
                    
        elif base1 == False and base2 == True  and base3 == True:  #2-3塁
            scoretmp = scoretmp + 2
            base1 = False
            base2 = False
            base3 = True
                    
        elif base1 == True  and base2 == True  and base3 == True:  #1-2-3塁
            scoretmp = scoretmp + 3
            base1 = False
            base2 = False
            base3 = True
            
    elif result == '2B':
        if base1 == False and base2 == False and base3 == False: #0塁
            scoretmp = scoretmp + 0
            base1 = False
            base2 = True
            base3 = False
                    
        elif base1 == True  and base2 == False and base3 == False: #1塁
            scoretmp = scoretmp + 0
            base1 = False
            base2 = True
            base3 = True
                    
        elif base1 == False and base2 == True  and base3 == False: #2塁
            scoretmp = scoretmp + 1
            base1 = False
            base2 = True
            base3 = False
                    
        elif base1 == False and base2 == False and base3 == True:  #3塁
            scoretmp = scoretmp + 1
            base1 = False
            base2 = True
            base3 = False
                    
        elif base1 == True  and base2 == True  and base3 == False: #1-2塁
            scoretmp = scoretmp + 1
            base1 = False
            base2 = True
            base3 = True
                    
        elif base1 == True  and base2 == False and base3 == True:  #1-3塁
            scoretmp = scoretmp + 1
            base1 = False
            base2 = True
            base3 = True
                    
        elif base1 == False and base2 == True  and base3 == True:  #2-3塁
            scoretmp = scoretmp + 2
            base1 = False
            base2 = True
            base3 = True
                    
        elif base1 == True  and base2 == True  and base3 == True:  #1-2-3塁
            scoretmp = scoretmp + 2
            base1 = False
            base2 = True
            base3 = True
                
    elif result == '1B' or 'FB' or 'DB':
        if base1 == False and base2 == False and base3 == False: #0塁
            scoretmp = scoretmp + 0
            base1 = True
            base2 = False
            base3 = False
                    
        elif base1 == True  and base2 == False and base3 == False: #1塁
            scoretmp = scoretmp + 0
            base1 = True
            base2 = True
            base3 = False
                    
        elif base1 == False and base2 == True  and base3 == False: #2塁
            scoretmp = scoretmp + 0
            base1 = True
            base2 = False
            base3 = True
                    
        elif base1 == False and base2 == False and base3 == True:  #3塁
            scoretmp = scoretmp + 1
            base1 = True
            base2 = False
            base3 = False
                    
        elif base1 == True  and base2 == True  and base3 == False: #1-2塁
            scoretmp = scoretmp + 0
            base1 = True
            base2 = True
            base3 = True
                    
        elif base1 == True  and base2 == False and base3 == True:  #1-3塁
            scoretmp = scoretmp + 1
            base1 = True
            base2 = True
            base3 = False
                    
        elif base1 == False and base2 == True  and base3 == True:  #2-3塁
            scoretmp = scoretmp + 1
            base1 = True
            base2 = False
            base3 = True
                    
        elif base1 == True  and base2 == True  and base3 == True:  #1-2-3塁
            scoretmp = scoretmp + 1
            base1 = True
            base2 = True
            base3 = True
            
    return base1, base2, base3, scoretmp
```

```python
# Set Information

hometeam = 'Yokohama'
awayteam = 'Seibu'
```

```python
# Set Data

hometeamfile = hometeam + '.csv'
awayteamfile = awayteam + '.csv'

with open(hometeamfile) as h:
    reader = csv.reader(h)
    l = [row for row in reader]
    
    for i in range(1,10):
        for j in range(0,10):
            homeplayersdata[i-1][j] = l[i][j]
            
with open(awayteamfile) as a:
    reader = csv.reader(a)
    l = [row for row in reader]
    
    for i in range(1,10):
        for j in range(0,10):
            awayplayersdata[i-1][j] = l[i][j]
```

```python
# Game

print('Game start:' + hometeam + ' VS ' + awayteam)

for inning in range(1,10): #1-9回
    
    if inningflg == 'top': 
        
        #表開始
        print('inning start ' + str(inning) + ' ' + inningflg)

        while outcount < 3:
            
            #打者設定
            batter = homeplayersdata[home_no]
            p_str = homeplayersdata[home_no][3:10]
            p = [float(i) for i in p_str]
            
            #打撃前表示
            print('打者:' + batter[1] + ':Before Batting:' + 'base1:' + str(base1), 'base2:' + str(base2), 'base3:' + str(base3), 'scoretmp:' + str(scoretmp), 'outcount:' + str(outcount))
            
            #打撃結果
            result = random.choices(resultlist, k = 1, weights = p)
            print(result[0])
            
            #結果状況
            if result[0] == 'OUT':
                outcount = outcount + 1
            else:
                base1, base2, base3, scoretmp = result_status(result[0], base1, base2, base3, scoretmp)
            
            #打撃後表示
            print('打者:' + batter[1] + ':After Batting:' + 'base1:' + str(base1), 'base2:' + str(base2), 'base3:' + str(base3), 'scoretmp:' + str(scoretmp), 'outcount:' + str(outcount))
                
            #打者交代
            if home_no < 8:
                home_no = home_no + 1
            else:
                home_no = 1
        
        #表終了
        score = scoretmp
        homescore[inning] = score
        print('inning end ' + str(inning) + ' ' + inningflg + ':' + str(homescore[inning]))
        
        #チェンジ
        score = 0
        scoretmp = 0
        outcount = 0
        base1 = False
        base2 = False
        base3 = False
        inningflg = 'bottom'
        
    if inningflg == 'bottom':
    
        #裏開始
        print('inning start ' + str(inning) + ' ' + inningflg)

        while outcount < 3:
            
            #打者設定
            batter = awayplayersdata[away_no]
            p_str = awayplayersdata[away_no][3:10]
            p = [float(i) for i in p_str]
            
            #打撃前表示
            print('打者:' + batter[1] + ':Before Batting:' + 'base1:' + str(base1), 'base2:' + str(base2), 'base3:' + str(base3), 'scoretmp:' + str(scoretmp), 'outcount:' + str(outcount))
            
            #打撃結果
            result = random.choices(resultlist, k = 1, weights = p)
            print(result[0])
            
            #結果状況
            if result[0] == 'OUT':
                outcount = outcount + 1
            else:
                base1, base2, base3, scoretmp = result_status(result[0], base1, base2, base3, scoretmp)
            
            #打撃後表示
            print('打者:' + batter[1] + ':After Batting:' + 'base1:' + str(base1), 'base2:' + str(base2), 'base3:' + str(base3), 'scoretmp:' + str(scoretmp), 'outcount:' + str(outcount))
                
            #打者交代
            if away_no < 8:
                away_no = away_no + 1
            else:
                away_no = 1

        #裏終了
        score = scoretmp
        awayscore[inning] = score
        print('inning end ' + str(inning) + ' ' + inningflg + ':' + str(awayscore[inning]))

        #チェンジ
        score = 0
        scoretmp = 0
        outcount = 0
        base1 = False
        base2 = False
        base3 = False
        inningflg = 'top'
        
    inning = inning + 1
```

```python
# Game Result

homescoresum = sum(homescore.values())
awayscoresum = sum(awayscore.values())

print('Home Score:', end='')
print(homescore)
print('Home Score Sum:', end='')
print(homescoresum)

print('Away Score:', end='')
print(awayscore)
print('Away Score Sum:', end='')
print(awayscoresum)

if homescoresum > awayscoresum:
    win = hometeam
    lose = awayteam
    
elif homescoresum < awayscoresum:
    win = awayteam
    lose = hometeam
    
else:
    win = 'No team'
    lose = 'No team'
    
print('Winner:' + win)
print('Loser:' + lose)
```
